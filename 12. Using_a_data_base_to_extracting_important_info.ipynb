```python
#IMPORT A PYTHON LIBRARY:
import os
import pandas as pd
import plotly.express as px
```

```python
#USING A METHOD THAT ALREADY EXISTS INSIDE THAT LIBRARY:

#THIS METHOD .listdir BRINGS ALL THE LIST OF ARCHIVES INSIDE A FOLDER:
arch_list = os.listdir('/content/drive/MyDrive/Curso Básico de Python/Base de dados/Vendas')
display(arch_list)
```

```python
#FOR LOOP:
for archive in arch_list:
  print(archive)
```

```python
#FOR LOOP TO IMPORT ARCHIVES:
for archive in arch_list:
  print(f'/content/drive/MyDrive/Curso Básico de Python/Base de dados/Vendas/{archive}')
```

```python
#PRINTING JUST ARCHIVES INSIDE THE CONDITION:
for archive in arch_list:
  if 'vendas' in archive.lower():
    print(f'/content/drive/MyDrive/Curso Básico de Python/Base de dados/Vendas/{archive}')
```

```python
for archive in arch_list:
  if 'vendas' in archive.lower():
    table = pd.read_csv(f'/content/drive/MyDrive/Curso Básico de Python/Base de dados/Vendas/{archive}')
    print(archive)
    display(table)
```

```python
complete_table = pd.DataFrame() #AN EMPTY TABLE TO START

for archive in arch_list:
  if 'vendas' in archive.lower():
    table = pd.read_csv(f'/content/drive/MyDrive/Curso Básico de Python/Base de dados/Vendas/{archive}')
    complete_table = complete_table.append(table) #ADDING THE CONTENT OF ALL TABLES IN THE NEW TABLE

display(complete_table)
```

```python
#CREATING A NEW TABLE WITH JUST THE DATA I NEED:
product_table = complete_table.groupby('Produto').sum() 
product_table = product_table[['Quantidade Vendida','Preco Unitario']].sort_values(by='Quantidade Vendida', ascending=False)
display(product_table)
```

```python
#ADDING A NEW COLUMN WITH DATA CALCULATED USING PREVIOUS DATA ALREADY IN THE TABLE:
complete_table['Total Billing'] = complete_table['Quantidade Vendida'] * complete_table['Preco Unitario']
display(complete_table)
```

```python
billing_table = complete_table.groupby('Produto').sum()
billing_table = billing_table[['Total Billing']].sort_values(by='Total Billing', ascending=False)
display(billing_table)
```

```python
shop_table = complete_table.groupby('Loja').sum()
shop_table = shop_table[['Total Billing']].sort_values(by='Total Billing', ascending=True)
display(shop_table)
```

```python
#GRAPHICS USING PLOTLY
graphic = px.bar(shop_table,x=shop_table.index,y='Total Billing')
graphic.show()
```
